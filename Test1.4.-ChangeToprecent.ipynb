# Fill NA as 0

```python
data = installment_payment

#NA will be assume not paid
data['PAY_DAYS'] = data['PAY_DAYS'].fillna(0)
data['AMT_PAY'] = data['AMT_PAY'].fillna(0)

data.head()
```

# Create new feature Late days and Amount left 

```python
#check if the SK_ID_PREV payment is late or no
data['LATE']=data['INST_DAYS']-data['PAY_DAYS']
#check if the amount paid is enough or no to pay AMT_INST
data['AMT_LEFT']=data['AMT_INST']-data['AMT_PAY']
data.head(100)
```

```python
data.sort_values(by='AMT_LEFT', ascending= True)
```

```python
data.sort_values(by='LATE', ascending=False)
```

# Change value to have load or not and late pay or not

```python
# #if the amount is more than enough thats mean the loan is paid, this variable means the LN_ID have bad 
#reputaion in terms of paying the amount
data['HAVE_LOAN'] = data['AMT_LEFT'].apply(lambda x: 0 if x <= 0 else 1) 

#if the late more than 30 thats mean he is late to pay the loan, this variable means the LN_ID have bad 
#reputaion in terms of the date of payment
data['LATE_PAYMENT'] = data['LATE'].apply(lambda x: 0 if x-30 < 0 else 1) 


data.head(100).sort_values(by='AMT_LEFT', ascending= True)
```

# create new feature based on the loan history

```python
data2=data.groupby('SK_ID_PREV').agg({'INST_NUMBER':['max'],'HAVE_LOAN':['mean'],'LATE_PAYMENT':['mean']})
data2
```

# Merge Prev_app and data 2, add 3 Inst number, Have load, and Late payment

```python
sec = pd.merge(left=prev_app, right=data2, how='left', left_on='SK_ID_PREV', right_on='SK_ID_PREV')
sec
```
